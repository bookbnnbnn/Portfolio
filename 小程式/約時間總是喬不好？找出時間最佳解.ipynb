```python
import sys
```

```python
D = {}
name_set = set()
while True: 
    name = input("請輸入名字：")    
    if name in name_set:
        print("這個名字已經登記過，將會覆蓋之前輸入的內容")
    name_set.add(name)  #將名字存入集合以判斷有沒有重複填寫
    while True:
        while True:
            try:
                week = int(input("星期幾有空(請輸入阿拉數字1~7)："))
                if week not in range(1, 8):
                    print("請輸入正確的時間")
                    continue
                else:
                    break
            except ValueError:
                print("請輸入正確的指令")
                continue
        while True:
            try:
                time_start = int(input("可以的時段起點(請輸入阿拉數字1~24)："))
                if time_start not in range(1, 25):
                    print("請輸入正確的時間")
                    continue
                else:
                    break
            except ValueError:
                print("請輸入正確的指令")
                continue            
        while True:
            try:
                time_end = int(input("可以的時段終點(請輸入阿拉數字1~24)："))
                if (time_end not in range(1, 25)) or (time_end <= time_start):
                    print("請輸入正確的時間")
                    continue
                else:
                    break
            except ValueError:
                print("請輸入正確的指令")
                continue
        while True:
            try:
                score = int(input("對此時段的喜好程度(請輸入阿拉數字1~5)："))
                if score not in range(1, 6):
                    print("請輸入正確的時間")
                    continue
                else:
                    break
            except ValueError:
                print("請輸入正確的指令")
                continue  
        time_list = []
        for i in range(time_start, time_end + 1):
            time_list.append(i)     #將起始和結束時間中間的整數填滿

        name_score_list = []        #將分數和名字放在一個list
        name_score_list.append(score)
        name_score_list.append(name)
        for i in time_list:         
            week_time = week, i     #將D這個字典的key用(week, time)這個tuple呈現
            if week_time in D.keys():   #當輸入的(week, time)原本就是D的key時
                if name not in D[week_time]:    #且如果名字不存在(week, time)的value
                    copy_list = D[week_time][:] #就在(week, time)的value後面貼上這個name並把分數相加
                    add_score = copy_list[0] + score
                    copy_list[0] = add_score                
                    copy_list.append(name)                
                    D[week_time] = copy_list
                else:
                    break
            else:
                D[week_time] = name_score_list  #當輸入的(week, time)原本不是D的key時，直接創一個
        E = D.copy()
        print("以下為剛剛輸入入的資料：")
        chinese_list = ["一", "二", "三", "四", "五", "六", "日"]
        print("%s可以的時段：星期%s%d~%d" %(name, chinese_list[week - 1], time_start, time_end))
        m0 = input("請確認資料是否正確(請輸入Y/N)：")
        while True:
            if (m0 == "n") or (m0 == "n".upper()):  #如果資料錯誤回到上一個輸入的時候
                D = E
                name_set.remove(name)
                break
            if (m0 == "y") or (m0 == "y".upper()):
                break
            else:
                print("請輸入正確的指令")
                m0 = input("請確認資料是否正確(請輸入Y/N)：")
                continue
        if (m0 == "n") or (m0 == "n".upper()):                
            break        

        
        m1 = input("是否還有另一個時段(請輸入Y/N)：")
        if (m1 == "n") or (m1 == "n".upper()):
            break
        if (m1 == "y") or (m1 == "y".upper()):
            continue
        else:
            while True:
                print("請輸入正確的指令")
                m1 = input("是否還有另一個時段(請輸入Y/N)：")
                if (m1 == "n") or (m1 == "n".upper()) or (m1 == "y") or (m1 == "y".upper()):
                    break
                else:
                    continue
            if (m1 == "n") or (m1 == "n".upper()):
                break
            if (m1 == "y") or (m1 == "y".upper()):
                continue


    m2 = input("是否繼續輸入另一個人資料(請輸入Y/N)：")
    if (m2 == "n") or (m2 == "n".upper()):
        break
    if (m2 == "y") or (m2 == "y".upper()):
        continue
    else:
        while True:
            print("請輸入正確的指令")
            m2 = input("是否繼續輸入另一個人資料(請輸入Y/N)：")
            if (m2 == "n") or (m2 == "n".upper()) or (m2 == "y") or (m2 == "y".upper()):
                break
            else:
                continue
        if (m2 == "n") or (m2 == "n".upper()):
            break
        if (m2 == "y") or (m2 == "y".upper()):
            continue


week_time_count = {}    #數看看哪一個(week, time)可以的人最多
L = []
for week_time, score_name in D.items(): 
    week_time_count[week_time] = len(score_name)    #將(week, time)的value放該時段可以的人數 
    L.append(len(score_name))   #將所有時段可以的時間放到list
chosen_week_time = []
for week_time, count in week_time_count.items():
    if count == max(L):     #如果該時段可以的人數是最大值就代表是最多人可以的時段
        chosen_week_time.append(week_time)



all_interval = []
for week_time in chosen_week_time:  #將時段和該時段的分數和可以的人合併成一個list方便運算
    all_interval += week_time
    all_interval.append(D[week_time])

#該list大約長成[week1, time1, [score1, name1], week2, time2, [score2, name2], ...]
chosen = {}
interval = []
for a in range(1, 8):
    for i in range(0, int(len(all_interval)/3 - 1)):    #該list是3個3個一組
        if 3*i % 3 == 0:
            if all_interval[3*i] == a:  #如果week一樣
                if all_interval[3*i+1] - all_interval[3*(i+1)+1] == -1: #時段相差一小時(代表是連續的時段)
                    interval.append(all_interval[3*i+1])    #將該時段貼上去
                    if i == int(len(all_interval)/3 - 2):   #當list最後一個跑完將它存成字典
                        interval.append(interval[-1] + 1)
                        interval.insert(0, all_interval[3*i])
                        interval = tuple(interval)
                        chosen[interval] = all_interval[3*i+2]
                else:
                    interval.append(interval[-1] + 1)   #或時段相差不為一小時的時候(不連續)將它存成字典
                    interval.insert(0, all_interval[3*i])
                    interval = tuple(interval)
                    chosen[interval] = all_interval[3*i+2]
                    interval = []
                    continue 


print("--------------------------------------")
i = 0
for score_name in chosen.values():
    if len(score_name) == 2:    #如果每個時段都只有一個人，定義成沒有大家可以的時段
        i += 1
if i == len(chosen.values()):
    print("沒有大家可以的時段")
    print("--------------------------------------")
    print("程式結束")
    sys.exit()
if len(chosen) == 0:    #如果選出來的時段是空的，也是大家都沒有的時段
    print(2)
    print("沒有大家可以的時段")
    print("--------------------------------------")
    print("程式結束")
    sys.exit()

score_list = []
for score_name in chosen.values():  #將所有分數取最大值
    score_list.append(score_name[0])
for week_time, score_name in chosen.items():    #如果分數等於最大值代表是大家綜合喜好程度最佳的時段
    week = week_time[0]
    time = list(week_time)
    time.pop(0)
    score = score_name[0]
    score_name.pop(0)
    name = score_name
    if score == max(score_list):
        chinese_list = ["一", "二", "三", "四", "五", "六", "日"]
        print("星期%s的%d~%d點" %(chinese_list[week - 1], min(time), max(time)))
        name_list = []
        for i in range(0, len(name)):
            name_list.append(name[i])
        print("可以去的人有%s" %",".join(name_list))
        print("--------------------------------------")
```
