```python
def youtube_search(search):
    from bs4 import BeautifulSoup
    import requests
    import numpy as np
    import re
    from IPython.display import display, YouTubeVideo
    url = f"http://m.youtube.com/results?search_query={search}"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    data = soup.find_all("", {"href":re.compile("^(/watch\?v=)[A-Za-z0-9]+$")})
    num = np.random.randint(0, len(data))
    key_word = data[num]["href"].split("=")[1] #選取爬到網頁的隨機一個
    myvideo = YouTubeVideo(key_word)
    display(myvideo)
```

```python
youtube_search("Yen-Lung Tsai")
```
