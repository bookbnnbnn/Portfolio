```python
def fugle_api_crawler(stock_id, api_token):
    import requests
    payload = {
        'symbolId':stock_id,
        'apiToken':api_token
    }
    res = requests.get('https://api.fugle.tw/realtime/v0/intraday/quote', params = payload)
    json_data = res.json()
    res1 = requests.get("https://api.fugle.tw/realtime/v0/intraday/meta", params=payload)
    json_data1 = res1.json()
    return{
        "priceHigh":json_data["data"]["quote"]["priceHigh"]["price"],
        "priceLow":json_data["data"]["quote"]["priceLow"]["price"],
        "priceOpen":json_data["data"]["quote"]["priceOpen"]["price"],
        "close": json_data["data"]["quote"]["trade"]["price"],
        "name":json_data1["data"]["meta"]["nameZhTw"],
        "priceReference":json_data1["data"]["meta"]["priceReference"]
    }
```

```python
import xlwings as xw
import time
while time.strftime("%H:%M:%S") <= '13:30:00': #在13:30收盤前都會執行
    wb = xw.Book(r"dashboard1.xlsx")
    sheet = wb.sheets['觀察清單']
    end_row = sheet.range('A1').end('down').row
    api_token = sheet.range('L1').value
    for i in range(2, end_row+1):
        stock_id = sheet.range(f'A{i}').value
        stock_id = str(int(stock_id)) if type(stock_id) == float else str(stock_id)
        print(stock_id)
        result = fugle_api_crawler(stock_id, api_token)
        print(result)
        print(list(result.keys()))
        sheet.range(f'B{i}').value = [
            result["name"], 
            result["priceOpen"],
            result["priceHigh"],
            result["priceLow"],
            result["close"],
            result["priceReference"],
            result["close"]/result["priceReference"] - 1
        ]
        # 漲是紅色，跌是綠色
        if sheet.range(f'H{i}').value < 0:
            sheet.range(f'H{i}').color = (0, 255, 0)
        elif sheet.range(f'H{i}').value > 0:
            sheet.range(f'H{i}').color = (255, 0, 0)
    sheet.range('I2').value = time.strftime("%Y/%m/%d %H:%M:%S")
    time.sleep(5)
# 下面的ERROR是因為我強制關掉他才會出現的
```
